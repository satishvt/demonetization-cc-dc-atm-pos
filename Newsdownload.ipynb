## Stage 2 of Data Collection
+ In this stage, news paper content was collected from the news links
+ For every news site, the content was stored in different divisions, classes. 
+ Using the selector gadget OR the Chrome developer tools/View Source - following details were extracted
    + Content
    + Date Published On
+ Collected Data was stored in a single excel for every news site

### Challenges Faced

+ Few sites maintain Date Publised and few other sites follow, Date Updated. So there can be some bias on this data
+ Formats of the dates used is not consistent
+ Few of the sites have news sites & blogs seperately, so the same news extraction logic wasn't usable
+ FirstPost site had java script code in the content, which had to be cleared manually

```python
### Constants & Imports
import bs4
from bs4 import BeautifulSoup
import requests
import urllib
import urllib3
import re
import os
import sys
import pandas as pd
from urllib.parse import urlparse
#import google
#Sentiment_Data_Folder ='/Users/apple/Gdrive/Term1/PRAT/Demonetization_Sentiment_Analysis/'
Sentiment_Data_Folder = 'D:/iCloudDrive/ISB/Term1/PRAT/Demonetization_Sentiment_Analysis/'
```

# The Hindu

```python
USER_AGENT = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
link_info = pd.DataFrame(columns=['site','link','date-published-updated'])

try:
    f_links = open(os.path.join(Sentiment_Data_Folder,"links.txt"),'r')
except:
    print("Links file not present")
    sys.exit()
i=0
for links in f_links:
    #print(links)
    t=requests.post(links)
    t1=BeautifulSoup(t.content.lower(),"lxml").find_all("div",id=re.compile("content-body",re.IGNORECASE))
    t3=BeautifulSoup(t.content,"lxml").find_all("none")
    parsed_uri = urlparse(links)

    for t4 in t3:
        datex=t4.text
        break
    #print(str(datex))
    link_info = link_info.append({'site':'The Hindu','link':links,'date-published-updated':str(datex).strip()},ignore_index=True)

    i = i+1
   
    if(os.path.isdir(os.path.join(Sentiment_Data_Folder,parsed_uri.netloc))):
        pass
    else:
        os.makedirs(os.path.join(Sentiment_Data_Folder,parsed_uri.netloc))
        
    f_write = open(os.path.join(os.path.join(Sentiment_Data_Folder,parsed_uri.netloc),str(i)+".txt"),"w",encoding='utf-8')
    for t2 in t1:
        f_write.write(t2.text)
    f_write.close()
    
#print(link_info['date-published-updated'])
link_info.to_csv(Sentiment_Data_Folder+"The-Hindu-Extract.csv")
```

## Firstpost

```python
USER_AGENT = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
link_info = pd.DataFrame(columns=['site','link','date-published-updated'])

try:
    f_links = open(os.path.join(Sentiment_Data_Folder,"links-firstpost.txt"),'r')
except:
    print("Links file not present")
    sys.exit()
i=0
for links in f_links:
    #print(links)
    t=requests.post(links)
    #print(t.status_code)
    t1=BeautifulSoup(t.content.lower(),"lxml").findAll("div",{"class":"article-full-content"})
    t3=BeautifulSoup(t.content,"lxml").find_all("span",class_="article-date")
    for t4 in t3:
        datex=t4.text.strip()
        break
    link_info = link_info.append({'site':'FirstPost','link':links,'date-published-updated':datex},ignore_index=True)
    i = i + 1
    #t1 = t1.encode("utf-8")
    parsed_uri = urlparse(links)
    
    if(os.path.isdir(os.path.join(Sentiment_Data_Folder,parsed_uri.netloc))):
        pass
    else:
        os.makedirs(os.path.join(Sentiment_Data_Folder,parsed_uri.netloc))
        
    f_write = open(os.path.join(os.path.join(Sentiment_Data_Folder,parsed_uri.netloc),str(i)+".txt"),"w",encoding='utf-8')
    for t2 in t1:
        f_write.write(t2.text)
    f_write.close()
    
link_info.to_csv(Sentiment_Data_Folder+"FirstPost-Extract.csv")        
```

# Times of India

```python
USER_AGENT = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
link_info = pd.DataFrame(columns=['site','link','date-published-updated'])

try:
    f_links = open(os.path.join(Sentiment_Data_Folder,"links-toi.txt"),'r')
except:
    print("Links file not present")
    sys.exit()
i=0
datex=0
for links in f_links:
    print(links.strip())
    t=requests.get(links.strip())
    print(t.status_code)
    print(t.reason)
    print(t.is_redirect)
    if(links.strip().endswith('.cms')):
        t1=BeautifulSoup(t.content.lower(),"lxml").find_all("div",class_="article_content clearfix")
        t3=BeautifulSoup(t.content,"lxml").find_all("span")
        for t4 in t3:
            if 'Updated:' in t4.text.strip():
                datex=t4.text.split(':')[1].strip()
    else:
        #rint("links-blog",links)
        t1=BeautifulSoup(t.content.lower(),"lxml").find_all("div",class_="content")
        t3=BeautifulSoup(t.content,"lxml").find_all("span",class_="date")
        for t4 in t3:
            datex=t4.text.strip()
            break
    #t1 = t1.encode("utf-8")
    

    
    link_info = link_info.append({'site':'Times of India','link':links.strip(),'date-published-updated':datex},ignore_index=True)
    i = i + 1
        
    parsed_uri = urlparse(links)
    
    if(os.path.isdir(os.path.join(Sentiment_Data_Folder,parsed_uri.netloc))):
        pass
    else:
        os.makedirs(os.path.join(Sentiment_Data_Folder,parsed_uri.netloc))
        
    f_write = open(os.path.join(os.path.join(Sentiment_Data_Folder,parsed_uri.netloc),str(i)+".txt"),"w",encoding='utf-8')
    for t2 in t1:
        f_write.write(t2.text)
    f_write.close()
    
link_info.to_csv(Sentiment_Data_Folder+"TOI-Extract.csv")        
```
